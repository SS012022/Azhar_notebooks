```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data={"Employee_id":[10,20,15,30,25],
      "Gender":['M','F','F','M','F'],
      "Remarks":['good','nice','good','great','good']}
df=pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
categorical_cols=df.select_dtypes(include=['object']).columns.tolist()
categorical_cols
```

```python
encoder=OneHotEncoder(sparse_output=False,drop="first")
one_hot_end=encoder.fit_transform(df[categorical_cols])
```

```python
df1=pd.DataFrame(one_hot_end,columns=encoder.get_feature_names_out(categorical_cols))
df1
```

```python
df1=pd.concat([df,df1],axis=1)
df1
```

```python
df1.drop(categorical_cols,axis=1,inplace=True)
df1
```

**Binary Encoding**

```python
data={"Employee_id":[10,25,15,30,24],
      "Gender":['M','F','F','M','F'],
      "Remarks":['good','nice','good','great','good']}
df=pd.DataFrame(data)
df
```

```python
x=np.mean(df["Employee_id"])
# I want the mean for M as mean of all M , and mean for F means mean of all F
mean_gender=df.groupby("Gender")["Employee_id"].mean()
mean_gender
```

```python
df["new_gender"]=df["Gender"].map(mean_gender)
df
```

```python
mean_remark=df.groupby("Remarks")["Employee_id"].mean()
df["new_Remarks"]=df["Remarks"].map(mean_remark)
df
```

```python
df.drop(["Gender","Remarks"],axis=1,inplace=True)
df
```
