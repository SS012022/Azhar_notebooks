```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(2,6,5)
y=np.linspace(10,14,5)
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y,c='r')
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.show()
```

```python
plt.plot(x,np.cos(x),marker='o')
plt.show()
```

```python
#x=np.linspace(2,6,10)
x=[2,5,7,9,12,16]
y=[i**2 for i in x]
y1=[i**.5 for i in x]
y2=[i**1.4 for i in x]
plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)

plt.show()
```

```python
day=[1,2,3,4,5]
study=[2,5,4,3,6]
play=[5,2,3,4,1]
#stacked graph
plt.legend()
plt.stackplot(day,study,play)
plt.show()
```

```python
day=[1,2,3,4,5]
study=[6,5,4,5,6]
play=[5,2,3,4,1]

plt.bar(day,study,color='g')
plt.bar(day,play,color='b')
plt.show()
```

```python
qty=[34,67,52,37,42]
items=['Mouse','Keyboard','Monitor','PenDrv','TFT']

plt.pie(qty,labels=items)
plt.show()
```

```python
df1=pd.DataFrame({'StuId':["sahil","saif","azhar"],'B':[10,20,30]})
df2=pd.DataFrame({'StuId':["sahil","saif","azhar"],'C':[21,22,23]})

merge_df=pd.merge(df1,df2,on='StuId')
print(merge_df)
values=merge_df.groupby('StuId')['B'].sum()
plt.xlabel("Student id")
plt.ylabel("Marks")
#plt.xticks(rotation=90)
values.plot(kind='bar')

plt.show()
```
