```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,10,2)
y=np.linspace(1,3,5)
plt.plot(x,y,marker='*',c='green',mec='pink',ms=10,mfc='pink')
plt.show()
```

```python
#subplots

x=np.linspace(2,5,6)
y=np.linspace(10,15,6)
plt.subplot(1,2,1)
plt.plot(x,y)

x1=np.linspace(2,5,6)
y1=np.linspace(12,14,6)
plt.subplot(1,2,2)
plt.plot(x1,y1)

plt.show()


```

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.legend()
plt.show()
```
